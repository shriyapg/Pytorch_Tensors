```python
import torch
my_torch=torch.arange(10)
# Reshape and View
my_torch=my_torch.reshape(2,5)
my_torch
# Error when reshape value as (2,6) since it can not have between 0 to 9.
```

```python
# Reshape if we don't know the number of items using -1
my_torch2=torch.arange(10)
my_torch2
```

```python
my_torch2=my_torch2.reshape(2,-1)
my_torch2
# my_torch2=torch.arange(15)
# my_torch2=my_torch2.reshape(2,-1)  ----> not work out but (3,-1) will work.
# (2,-1) can be written as (-1,5)
```

```python
import torch
my_torch3=torch.arange(10)
my_torch3
```

```python
# Slices ---> It allow you to grab a section of your tensor
my_torch7=torch.arange(10)
my_torch7
```

```python
# Grab a specific item
my_torch7[7]
```

```python
# Grab slice
my_torch8=my_torch7.reshape(5,2)
my_torch8
```

```python
my_torch8[:,1]
```

```python
# Return column
my_torch8[:,1:]
```
